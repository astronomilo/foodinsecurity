```python
# Packages
import pandas as pd
```

```python
# Read dataset
path = './datasets/CPS2020/'
csvfile = 'dec20pub.csv'
df = pd.read_csv(path + csvfile)
```

```python
df.head()
```

**We want to select only data for Mississipi**. According to page 6-9 in **cpsdec20.pdf**, we can use the Federal Iinformation Processing Standards (FIPS) state code to accomplish this task. We use the `GESTFIPS` column and look for the number `28`, which corresponds to MS (Mississipi).

```python
ms = 28
df_ms = df.loc[df.GESTFIPS == ms]
df_ms
```

```python
# Save dataframe to a new CSV file containing only data for Mississipi
df_ms.to_csv('./datasets/Mississipi/CPS_2020Data_MS.csv')
```
