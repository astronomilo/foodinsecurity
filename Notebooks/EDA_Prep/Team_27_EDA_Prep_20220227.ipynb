#### How the Income and food expenses correlate with the number of children in the house?

```python
import warnings
warnings.filterwarnings("ignore")

def fix_number_children(df, string):
    for i in np.arange(0,3):
    # for i in np.arange(0,7):
        if i not in df.NUMBER_OWN_CHILDREN.drop_duplicates().values:
            df = df.append({'NUMBER_OWN_CHILDREN': i,
                             string: 0}, 
                             ignore_index = True)
    return df

# Chose number of children with statitistical significance
Ch_cps2019_df = cps2019_df[cps2019_df.NUMBER_OWN_CHILDREN.isin([0,1,2,3])]

# Households needing to meet food needs
FoodNeed = Ch_cps2019_df[Ch_cps2019_df['AMOUNT_NEEDED_TO_MEET_HOUSEHOLD_FOOD_NEED']!=-1]
# Add a zero to those not defined values of NUMBER_OWN_CHILDREN
FoodNeed = fix_number_children(FoodNeed, 'AMOUNT_NEEDED_TO_MEET_HOUSEHOLD_FOOD_NEED')

# Food stamps/month:
FoodStamps = Ch_cps2019_df[Ch_cps2019_df['FOODSTAMPS_AMT_PERMONTH']>0]
# Add a zero to those not defined values of NUMBER_OWN_CHILDREN
FoodStamps = fix_number_children(FoodStamps, 'FOODSTAMPS_AMT_PERMONTH')

# Purchases in Restaurants /week
FoodRest = Ch_cps2019_df[Ch_cps2019_df['HOUSEHOLD_PURCHASES_RESTAURANTS']>=0]
FoodRest = fix_number_children(FoodRest, 'HOUSEHOLD_PURCHASES_RESTAURANTS')

# Purchases in Groceries / week
FoodGroc = Ch_cps2019_df[Ch_cps2019_df['HOUSEHOLD_PURCHASES_GROCERIES']>=0]
FoodGroc = fix_number_children(FoodGroc, 'HOUSEHOLD_PURCHASES_GROCERIES')

# Family Income
FamInc = Ch_cps2019_df[Ch_cps2019_df['AMOUNT_NEEDED_TO_MEET_HOUSEHOLD_FOOD_NEED']!=-1]
Dict_Family_Income = { 1: 2500, 2: 6250, 3: 8750, 4: 11250, 5: 13750, 6: 17500,
                       7: 22500, 8: 27500, 9: 32500, 10: 37500, 11: 45000, 12: 55000,
                      13: 67500, 14: 87500, 15: 125000, 16: 150000}
FamInc = FamInc.replace({"FAMILY_INCOME": Dict_Family_Income})
FamInc = fix_number_children(FamInc, 'FAMILY_INCOME')

# MAKE THE PLOT
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, sharex=True, figsize=(12,15))
fig.subplots_adjust(0,0,1,1,0,0.03) # This removes space between plots

# Money needed/week:
FoodNeed.boxplot(column='AMOUNT_NEEDED_TO_MEET_HOUSEHOLD_FOOD_NEED',
                 by='NUMBER_OWN_CHILDREN',grid=False, ax=ax1)
ax2.set_xticklabels(ax1.get_xticklabels(),ha='left')
ax1.set_ylabel('Money needed/week', fontsize=12)
ax1.set_title('')
ax1.set_ylim(0,210)
ax1.axes.xaxis.set_ticklabels([])
# Food stamps/month:
FoodStamps.boxplot(column='FOODSTAMPS_AMT_PERMONTH',
                 by='NUMBER_OWN_CHILDREN',grid=False, ax=ax2)
ax2.set_ylabel('Food stamps/month', fontsize=12)
ax2.set_title('')
ax2.set_ylim(0,700)
ax2.axes.xaxis.set_ticklabels([])
# Purchase Restaurants
FoodRest.boxplot(column='HOUSEHOLD_PURCHASES_RESTAURANTS',
                 by='NUMBER_OWN_CHILDREN',grid=False,ax=ax3)
ax3.set_ylabel('Money Restaurants/week', fontsize=12)
ax3.set_title('')
ax3.set_ylim(0,305)
ax3.axes.xaxis.set_ticklabels([])

# Purchases in Groceries / week
FoodGroc.boxplot(column='HOUSEHOLD_PURCHASES_GROCERIES',
                 by='NUMBER_OWN_CHILDREN',grid=False,ax=ax4)
ax4.set_ylabel('Money Groceries/week', fontsize=12)
ax4.set_title('')
ax4.set_ylim(0,450)
ax4.axes.xaxis.set_ticklabels([])

# Family Income
FamInc.boxplot(column='FAMILY_INCOME',
                 by='NUMBER_OWN_CHILDREN',grid=False, ax=ax5)
ax5.set_ylabel('Family Income/year', fontsize=12)
ax5.set_title('')
ax5.set_ylim(0,160000)
ax5.set_xlabel('Number of children under 18YO in household')

fig.suptitle('2019',y=0.93, fontsize=30)
plt.show()
```

```python
Dict_Race = {1: 'White Only',
             2: 'Black Only',
             3: 'American Indian, Alaskan Native Only',
             4: 'Asian Only',
             5: 'Hawaiian/Pacific Islander Only',
             6: 'White-Black',
             7: 'White-AI',
             8: 'White-Asian',
            16: 'W-B-AI',
            21: 'W-A-HP'}

def fix_race(df, string):
    for i in [1,2,4,6]:
    #for i in [1,2,3,4,5,6,7,8,16,21]:
        if i not in df.RACE.drop_duplicates().values:
            df = df.append({'RACE': i,
                             string: 0}, 
                             ignore_index = True)
    df = df.replace({"RACE": Dict_Race})    
    return df

# Chose Races with statitistical significance
Rz_cps2019_df = cps2019_df[cps2019_df.RACE.isin([1,2,4,6])]

# Households needing to meet food needs
FoodNeed = Rz_cps2019_df[Rz_cps2019_df['AMOUNT_NEEDED_TO_MEET_HOUSEHOLD_FOOD_NEED']!=-1]
FoodNeed = fix_race(FoodNeed, 'AMOUNT_NEEDED_TO_MEET_HOUSEHOLD_FOOD_NEED')
# Food stamps/month:
FoodStamps = Rz_cps2019_df[Rz_cps2019_df['FOODSTAMPS_AMT_PERMONTH']>0]
FoodStamps = fix_race(FoodStamps, 'FOODSTAMPS_AMT_PERMONTH')
# Purchases in Restaurants /week
FoodRest = Rz_cps2019_df[Rz_cps2019_df['HOUSEHOLD_PURCHASES_RESTAURANTS']>=0]
FoodRest = fix_race(FoodRest, 'HOUSEHOLD_PURCHASES_RESTAURANTS')
# Purchases in Groceries / week
FoodGroc = Rz_cps2019_df[Rz_cps2019_df['HOUSEHOLD_PURCHASES_GROCERIES']>=0]
FoodGroc = fix_race(FoodGroc, 'HOUSEHOLD_PURCHASES_GROCERIES')
# Family Income
FamInc = Rz_cps2019_df[Rz_cps2019_df['AMOUNT_NEEDED_TO_MEET_HOUSEHOLD_FOOD_NEED']!=-1]
Dict_Family_Income = { 1: 2500, 2: 6250, 3: 8750, 4: 11250, 5: 13750, 6: 17500,
                       7: 22500, 8: 27500, 9: 32500, 10: 37500, 11: 45000, 12: 55000,
                      13: 67500, 14: 87500, 15: 125000, 16: 150000}
FamInc = FamInc.replace({"FAMILY_INCOME": Dict_Family_Income})
FamInc = fix_race(FamInc, 'FAMILY_INCOME')

# MAKE THE PLOT
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, sharex=True, figsize=(12,15))
fig.subplots_adjust(0,0,1,1,0,0.03) # This removes space between plots

FoodNeed.boxplot(column='AMOUNT_NEEDED_TO_MEET_HOUSEHOLD_FOOD_NEED',
                 by='RACE',grid=False, ax=ax1)
ax1.set_ylabel('Money needed/week', fontsize=12)
ax1.set_title('')
ax1.set_ylim(0,210)
# Food stamps/month:
FoodStamps.boxplot(column='FOODSTAMPS_AMT_PERMONTH',
                 by='RACE',grid=False, ax=ax2)
ax2.set_ylabel('Food stamps/month', fontsize=12)
ax2.set_title('')
ax2.set_ylim(0,700)
# Purchase Restaurants
FoodRest.boxplot(column='HOUSEHOLD_PURCHASES_RESTAURANTS',
                 by='RACE',grid=False,ax=ax3)
ax3.set_ylabel('Money Restaurants/week', fontsize=12)
ax3.set_title('')
ax3.set_ylim(0,305)
# Purchases in Groceries / week
FoodGroc.boxplot(column='HOUSEHOLD_PURCHASES_GROCERIES',
                 by='RACE',grid=False,ax=ax4)
ax4.set_ylabel('Money Groceries/week', fontsize=12)
ax4.set_title('')
ax4.set_ylim(0,450)
# Family Income
FamInc.boxplot(column='FAMILY_INCOME',
                 by='RACE',grid=False, ax=ax5)
ax5.set_ylabel('Family Income/year', fontsize=12)
ax5.set_title('')
ax5.set_ylim(0,160000)
ax5.set_xlabel('Race')
ax5.tick_params(axis='x', labelrotation = 90)

fig.suptitle('2019',y=0.93, fontsize=30)
plt.show()
```
