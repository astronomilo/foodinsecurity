# CPS2020

```python
# Packages
import pandas as pd
```

```python
# Read dataset
path = '../../datasets/CPS/'
csvfile = 'dec20pub.csv'
df = pd.read_csv(path + csvfile)
```

```python
df.head()
```

**We want to select only data for Mississipi**. According to page 6-9 in **cpsdec20.pdf**, we can use the Federal Iinformation Processing Standards (FIPS) state code to accomplish this task. We use the `GESTFIPS` column and look for the number `28`, which corresponds to MS (Mississipi).

```python
ms = 28
df_ms = df.loc[df.GESTFIPS == ms]
df_ms
```

```python
# All_keys = ['HRYEAR4','GTCO','GTCBSASZ','GTCSA',
#         'HEFAMINC','HRHTYPE','PTDTRACE','PREMPNOT','PRCHLD','HETS2O','HETS6O','HES8B',
#         'HETS8CO','HETS8DO','HES9','HESP1','HESP21','HETSP3O','HESP6','HESP7','HESP7A',
#         'HESP8','HETSP9','HESS1','HESS2','HESS3','HESS4','HESH2','HESHF2','HESHM2','HESS5',
#         'HESS6','HESH1','HESSH2']

keys = ['HEFAMINC','HRHTYPE','PTDTRACE','PREMPNOT','PRCHLD',
        'HETS2O','HETS6O','HETS8CO','HETSP3O','HESP1']
```

```python
df_ms = df_ms[df_ms.columns.intersection(keys)]
df_ms
```

```python
# Save dataframe to a new CSV file containing only data for Mississipi
df_ms.to_csv('../../datasets/Mississipi/CPS_2020Data_MS.csv')
```
